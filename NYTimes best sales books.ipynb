# The New York Times Best Sellers

## books in all genre and between 2009/01/01 and 2017/01/01

### imports packages

```python
import requests
import pandas as pd
import os
```

### get json data from URL

```python
def get_raw_data(request):
    raw_data = requests.get(request).json()
    print("get json file from url : {}".format(request))
    print("--------------------------------------------")
    return raw_data
```

### make data frame from json file that extract from URL

```python
def make_Data_Frame(raw_data):
    frames = []

    for items in raw_data['results']['lists']:
        for books in items['books']:
            df = pd.DataFrame.from_dict(books)
            frames.insert(len(frames),df)
        
    results = pd.concat(frames)
    return(results)
```

### get excel file from Data Frame

```python
def export_excel(path,data_frame):
    writer = pd.ExcelWriter(path)
    data_frame.to_excel(writer,'Sheet1')
    writer.save()
    print("export excel from Dataframe & save in dir : {}".format(path))
```

### set api key that get from http://developer.nytimes.com/

```python
api_key = ""
```

### make list of data frame from different years of 2009 to 2017

```python
frames = []
```

```python
for i in range (9,18):
    request = "https://api.nytimes.com/svc/books/v3/lists/overview.json?api-key={}&published_date={}-01-01".format(api_key,str(2000 + i))
    print("url : " + request)
    print()
    raw_data = get_raw_data(request)
    df = make_Data_Frame(raw_data)
    
    frames.insert(len(frames),df)
```

### concat items of list of data frames to make one data frame

```python
results = pd.concat(frames)
results
```

### set path to export excel file 

```python
path = ""

if not os.path.exists(path):
    export_excel(path,results)
else:
    print("path exists!")
```
